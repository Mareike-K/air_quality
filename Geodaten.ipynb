```python
import json
import pandas as pd

# JSON-Datei laden
with open('data/airquality-covid19-cities.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

data

```

```python
data = data["data"] 

# print(type(data))  # Sollte jetzt <class 'list'> ausgeben
# print(len(data))   # Zeigt die Anzahl der Städte
# print(data[0])     # Zeigt die Struktur des ersten Eintrags
```

```python
# Erstellen eines DataFrames mit Städten und Geokoordinaten
df_places = pd.DataFrame([
    {
        "City": entry["Place"]["name"],  # Stadtname
        "Latitude": entry["Place"]["geo"][0],  
        "Longitude": entry["Place"]["geo"][1]
    }
    for entry in data if "Place" in entry  # Nur Einträge mit "Place" verwenden
])

# DataFrame anzeigen
df_places.head()
```

```python
# Standardisiere den Stadtnamen für eine bessere Übereinstimmung
df_air_quality["City"] = df_air_quality["City"].str.lower().str.strip()
df_places["City"] = df_places["City"].str.lower().str.strip()

# Zusammenführen der beiden DataFrames über "City"
df_combined = df_air_quality.merge(df_places, on="City", how="left")

df_combined.head()
```

```python
df_geodata = df_combined[["City", "Latitude", "Longitude"]].drop_duplicates()
df_geodata.head()
```

```python
import geopandas as gpd
import geodatasets
import matplotlib.pyplot as plt

# Weltkarte laden
world = gpd.read_file(geodatasets.get_path("naturalearth.land"))

# Plot erstellen
fig, ax = plt.subplots(figsize=(10, 6))

# Umrisse der Weltkarte plotten
world.plot(ax=ax, color='white', edgecolor='black')

# Städte als Scatterplot hinzufügen
ax.scatter(df_combined["Longitude"], df_combined["Latitude"], color='red', marker='o', s=4, label="Städte")

# Achsentitel setzen
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Städte mit Luftqualitätsmessungen")

# Legende hinzufügen
ax.legend()

# Plot anzeigen
plt.show()
```

```python
df_europe = df_combined[
    (df_combined["Latitude"] >= 35) & (df_combined["Latitude"] <= 70) &  # Breitengrade für Europa
    (df_combined["Longitude"] >= -25) & (df_combined["Longitude"] <= 50)  # Längengrade für Europa
]

df_europe
```

```python
# Natural Earth Daten von offizieller Quelle laden
url = "https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson"
world = gpd.read_file(url)

# Prüfen, welche Spalten enthalten sind
print(world.columns)

# Nur Europa auswählen (falls "continent" nicht existiert, evtl. "region_un", "subregion")
if "continent" in world.columns:
    europe = world[world["continent"] == "Europe"]
else:
    europe = world[world["region_un"] == "Europe"]

# Plot erstellen
fig, ax = plt.subplots(figsize=(10, 6))

# Umrisse der Europakarte plotten
europe.plot(ax=ax, color='white', edgecolor='black')

# Städte als Scatterplot hinzufügen
ax.scatter(df_europe["Longitude"], df_europe["Latitude"], color='red', marker='o', s=10, label="Städte in Europa")

# Achsentitel setzen
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Städte mit Luftqualitätsmessungen")

# Legende hinzufügen
ax.legend()

# Plot anzeigen
plt.show()
```
