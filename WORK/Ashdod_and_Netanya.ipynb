```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
```

```python
# Settings for displaying floats
pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
df = pd.read_csv("../data/cleaned_data.csv")
df.head()
```

```python
df = df.drop(columns=['Temperature', 'Wind-speed', 'Pressure'])
```

```python
df["Date"] = pd.to_datetime(df[["Year", "Month", "Day"]])
```

```python
# Alle verfügbaren Messmonate für Ashdod
df[df["City"] == "Ashdod"]["Date"].min(), df[df["City"] == "Ashdod"]["Date"].max()

# Anzahl Messwerte pro Monat
df[df["City"] == "Ashdod"].groupby("Month")["Co"].count()

# Tagesverlauf prüfen
df[df["City"] == "Ashdod"].groupby("Date")["Co"].mean().plot()
```

```python
df_ashdod_co = df[(df["City"] == "Ashdod")][["Date", "Co"]]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(df_ashdod_co["Date"], df_ashdod_co["Co"], color="red", label="Messwerte")

plt.xlabel("Datum")
plt.ylabel("CO-Konzentration")
plt.title("CO-Messwerte in Ashdod über die Zeit")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

Die Messwerte für Ashdod sind nur für einzelne Zeiträume vorhanden und sehr volatil. Ashdod wird aus den weiteren Analysen ausgeschlossen. Siehe NB "Ashkelon für mehr Details.
